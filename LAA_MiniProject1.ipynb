```python
# Step 1: Load real-world data

import numpy as np
import pandas as pd

from google.colab import files
uploaded = files.upload()

# Read dataset
df = pd.read_csv('student_marks1.csv')

# Show data
df
```

```python
# Step 2: Convert data into matrix form

# Remove non-numeric column
data = df.drop(columns=['Student'])

# Convert to numpy matrix
A = data.to_numpy()

print("Matrix A:")
print(A)
```

```python
# Step 3: Matrix simplification using RREF

from sympy import Matrix

M = Matrix(np.nan_to_num(A))  # replace NaN for computation
rref_matrix, pivots = M.rref()

print("RREF Matrix:")
print(rref_matrix)
```

```python
# Step 4: Analyze vector space

rank = np.linalg.matrix_rank(np.nan_to_num(A))
nullity = A.shape[1] - rank

print("Rank:", rank)
print("Nullity:", nullity)
```

```python
# Step 5: Remove redundant features

# We approximate this using rank info
print("Number of independent features:", rank)
```

```python
# Step 6: Orthogonalization using QR decomposition

Q, R = np.linalg.qr(np.nan_to_num(A))

print("Orthogonal basis (Q):")
print(Q)
```

```python
# Step 7: Projection of a subject's scores

b = np.nan_to_num(A[:, 0])  # first subject's scores

projection = Q @ Q.T @ b

print("Projection of first subject's scores:")
print(projection)
```

```python
# Step 8: Predict missing values using Least Squares

A_original = data.to_numpy()
A_filled = A_original.copy()

rows, cols = A_original.shape

for col in range(cols):

    missing_rows = np.isnan(A_original[:, col])

    if np.sum(missing_rows) == 0:
        continue

    b = A_original[:, col]
    A_reduced = np.delete(A_original, col, axis=1)

    valid_rows = ~np.isnan(b)

    A_train = A_reduced[valid_rows]
    b_train = b[valid_rows]

    A_train = np.nan_to_num(A_train)

    x = np.linalg.lstsq(A_train, b_train, rcond=None)[0]

    A_test = A_reduced[missing_rows]
    A_test = np.nan_to_num(A_test)

    predictions = A_test @ x

    A_filled[missing_rows, col] = predictions

    print(f"\nPredictions for {data.columns[col]}:")
    for idx, val in zip(np.where(missing_rows)[0], predictions):
        print(f"Student {idx+1}: {val:.2f}")
```

```python
# Step 9: Pattern discovery using eigenvalues

cov_matrix = np.cov(np.nan_to_num(A).T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:")
print(eigenvalues)
```

```python
# Step 10: Diagonalization

D = np.diag(eigenvalues)

print("Diagonal Matrix D:")
print(D)

print("Eigenvector Matrix:")
print(eigenvectors)
```

```python
# Step 11: Final output

print("Final Completed Matrix (with predicted values):")
print(A_filled)
```
